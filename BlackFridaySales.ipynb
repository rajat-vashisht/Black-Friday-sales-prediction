```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('train.csv')
```

```python
dataset.head()
```

```python
dataset['Age'].unique()
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
dataset['Product_Category_2'].fillna(dataset['Product_Category_2'].median(),inplace=True)
dataset['Product_Category_3'].fillna(dataset['Product_Category_3'].median(),inplace=True)
```

```python
dataset.head()
```

```python
dataset['Stay_In_Current_City_Years'].unique()
```

```python
dataset['Stay_In_Current_City_Years']=dataset['Stay_In_Current_City_Years'].apply(lambda x:str(x).replace('+','') if '+' in str(x) else str(x))
dataset['Stay_In_Current_City_Years']=dataset['Stay_In_Current_City_Years'].apply(lambda x:float(x))
```

```python
dataset.info()
```

```python
gender_dummy=pd.get_dummies(dataset['Gender'],drop_first=True)
age_dummy=pd.get_dummies(dataset['Age'],drop_first=True)
city_dummy=pd.get_dummies(dataset['City_Category'],drop_first=True)
```

```python
dataset=pd.concat([dataset,gender_dummy,age_dummy,city_dummy],axis=1)
```

```python
dataset.drop(['Gender','Age','City_Category'],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
X=dataset.drop(['User_ID','Product_ID','Purchase'],axis=1)
```

```python
y=dataset['Purchase']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
rf=RandomForestRegressor(n_estimators=10,random_state=0)
```

```python
rf.fit(X,y)
```

```python
test_data=pd.read_csv('test.csv') 
```

```python
test_data.head()
```

```python
test_data.isnull().sum()
```

```python
test_data['Product_Category_2'].fillna(test_data['Product_Category_2'].median(),inplace=True)
test_data['Product_Category_3'].fillna(test_data['Product_Category_3'].median(),inplace=True)
```

```python
test_data.isnull().sum()
```

```python
final_data=test_data.drop(['User_ID','Product_ID'],axis=1)
```

```python
final_data.head()
```

```python
final_data['Stay_In_Current_City_Years']=final_data['Stay_In_Current_City_Years'].apply(lambda x:str(x).replace('+','') if '+' in str(x) else str(x))
final_data['Stay_In_Current_City_Years']=final_data['Stay_In_Current_City_Years'].apply(lambda x:float(x))
```

```python
final_data.head()
```

```python
final_data.info()
```

```python
gender_dummy1=pd.get_dummies(final_data['Gender'],drop_first=True)
age_dummy1=pd.get_dummies(final_data['Age'],drop_first=True)
city_dummy1=pd.get_dummies(final_data['City_Category'],drop_first=True)
```

```python
final_data=pd.concat([final_data,gender_dummy1,age_dummy1,city_dummy1],axis=1)
```

```python
final_data.drop(['Gender','Age','City_Category'],axis=1,inplace=True)
```

```python
final_data.head()
```

```python
y_predicted=rf.predict(final_data)
```

```python
submission = pd.DataFrame({
        "Purchase": y_predicted,
        "User_ID": test_data['User_ID'],
        "Product_ID": test_data['Product_ID']
    })
```

```python
submission.to_csv('Blackfridaysales_submission.csv', index=False)
```
